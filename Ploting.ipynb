```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('AirQualityUCI.csv')
```

```python
df
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sn.boxplot(data=df)
plt.title('Boxplot of Air Quality Variables')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['CO(GT)'], bins=20, alpha=0.7, color='blue', edgecolor='black', linewidth=1.5, density=True)
plt.title('Histogram of CO(GT)')
plt.xlabel('CO(GT)')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['CO(GT)'], df['NO2(GT)'], marker='o', linestyle='-')
plt.title('NO2(GT) Over Time')
plt.xlabel('co')
plt.ylabel('NO2(GT)') 
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['CO(GT)'], marker='o', linestyle='-', label='CO(GT)')
plt.plot(df.index, df['NOx(GT)'], marker='o', linestyle='-', label='NOx(GT)')
plt.plot(df.index, df['NO2(GT)'], marker='o', linestyle='-', label='NO2(GT)')
plt.title('Air Quality Variables Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
